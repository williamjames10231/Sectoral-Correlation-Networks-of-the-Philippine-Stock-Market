```python
from itertools import combinations
from sage.all import Graph
from sage.all import matrix
from sage.all import RDF
import matplotlib.pyplot as plt

import pandas as pd
from Correlations_Distance_matrices_engine.data_loader import DataLoader
from Correlations_Distance_matrices_engine.correlation_distance_matrix_engine import CorrelationDistanceMatrixEngine
```

### Setup
## Loading Data and matrix computations

```python
loader_data = DataLoader()
engine_computations = CorrelationDistanceMatrixEngine(loader_data)
engine_splits = engine_computations.splitter()
batch_corr = [pd.DataFrame(), pd.DataFrame()]
batch_dist = [pd.DataFrame(), pd.DataFrame()]
batch_corr[0], batch_corr[1], batch_dist[0], batch_dist[1] = engine_computations.make_yields_splitter(engine_splits)
```

### Graph Visualizations
## Baseline Complete Graph

```python
sectors = ["Property", "Mining & Oil", "Financials", "Holding Firms", "Industrial", "Services"]
K6 = Graph()
K6.add_vertices(sectors)
K6.add_edges(combinations(sectors, 2))

for u, v, _ in K6.edges():
    K6.set_edge_label(u,v, "{:.2f}".format(0.00))

K6.show(figsize=8, edge_labels=True, layout='circular')
K6.plot(figsize=8, edge_labels=True, layout='circular').save("CompleteGraph.png")
```

![Complete_Graph](CompleteGraph.png)

## Correlations weighted Complete Graph
# batch a

```python
corr_matrix_batch_a = matrix(RDF,6,6, batch_corr[0].values)
K6_batch_a_corr = Graph(corr_matrix_batch_a, format="adjacency_matrix", vertex_labels=sectors)
K6_batch_a_corr.remove_loops()
mapping_a = {i: sectors[i] for i in range(6)}
K6_batch_a_corr.relabel(mapping_a)
K6_batch_a_corr.show(figsize=8, edge_labels=True, layout='circular')
K6_batch_a_corr.plot(figsize=8, edge_labels=True, layout='circular').save("CompleteGraph_batch_a_corr.png")
```

![batch_a_corr](CompleteGraph_batch_a_corr.png)

## Minimum Spanning Tree via Kruskal's Algorithm
# batch a

```python
K6_batch_a_dist_copy = K6_batch_a_dist
K6_batch_a_dist_copy_mst_edges = K6_batch_a_dist_copy.min_spanning_tree(algorithm="Kruskal")
K6_batch_a_dist_mst = Graph(K6_batch_a_dist_copy_mst_edges,weighted=True)
K6_batch_a_dist_mst.show(figsize=8, edge_labels=True, layout='spring', edge_color='red')
K6_batch_a_dist_mst.plot(figsize=8, edge_labels=True, layout='spring').save("CompleteGraph_batch_a_dist_mst.png")
```

![batch_a_mst](CompleteGraph_batch_a_dist_mst.png)

# batch b

```python
K6_batch_b_dist_copy = K6_batch_b_dist
K6_batch_b_dist_copy_mst_edges = K6_batch_b_dist_copy.min_spanning_tree(algorithm="Kruskal")
K6_batch_b_dist_mst = Graph(K6_batch_b_dist_copy_mst_edges,weighted=True)
K6_batch_b_dist_mst.show(figsize=8, edge_labels=True, layout='spring', edge_color='red')
K6_batch_b_dist_mst.plot(figsize=8, edge_labels=True, layout='spring').save("CompleteGraph_batch_b_dist_mst.png")
```
